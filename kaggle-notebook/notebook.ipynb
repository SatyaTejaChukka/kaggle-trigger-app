```python
import os
from datetime import datetime
import platform
```

## Folder inside Kaggle working directory

```python

folder_path = "/kaggle/working/runtime_logs"
os.makedirs(folder_path, exist_ok=True)
```

## Get current timestamp

```python

now = datetime.now()
timestamp = now.strftime("%Y-%m-%d_%H-%M-%S")
```

## Print runtime timestamp

```python

print("Notebook run at:", timestamp)
```

## Create filename with timestamp

```python

filename = f"run_{timestamp}.txt"
file_path = os.path.join(folder_path, filename)
```

## Write to file

```python
with open(file_path, "w") as f:
    f.write(f"Notebook executed at: {timestamp}\n")
    f.write(f"Python version: {platform.python_version()}\n")
    f.write(f"System: {platform.system()} {platform.release()}\n")

print("Run file created:", file_path)
```

## Maintain a cumulative history log

```python

history_file = os.path.join(folder_path, "run_history.txt")

with open(history_file, "a") as f:
    f.write(f"{timestamp}\n")

print("History updated:", history_file)
```

## List files created

```python
print("\nFiles currently in runtime_logs:")
print(os.listdir(folder_path))
```
