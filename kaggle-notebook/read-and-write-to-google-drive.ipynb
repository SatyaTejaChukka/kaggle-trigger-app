```python
file_path = "dummy.csv"

print("Entered file name: ", file_path)
```

# Reading the original file

```python
import pandas as pd

df = pd.read_csv(file_path)
print("Original Data:")
print(df)
```

# Making some changes

```python
df["score"] = [10, 20, 30]
df["status"] = ["new", "old", "new"]

# ➕ Add new rows
new_rows = pd.DataFrame({
    "id": [4, 5],
    "value": ["d", "e"],
    "score": [40, 50],
    "status": ["old", "new"]
})
```

# Concatenating

```python
df_new = pd.concat([df, new_rows], ignore_index=True)

print("\nUpdated Data:")
print(df_new)
```

```python
from datetime import datetime

now = datetime.now()
timestamp = now.strftime("%Y-%m-%d_%H-%M-%S")
new_file_name = f"updated_data_at_{timestamp}.csv"
df_new.to_csv(new_file_name, index=False)

print("\nNew CSV saved:", new_file_name)
```

# **Final Step:** Upload to destination folder

```python

gfile = drive.CreateFile({
    'title': new_file_name,
    'parents': [{'id': destination_folder_id}]
})

gfile.SetContentFile(new_file_name)
gfile.Upload()

print("\nFile uploaded to destination folder!")
```
